#### Task 5.2 Perform the Analysis (20 points)

- Perform the proposed analysis.
- Be sure to comment your code throughout so that it is easy to follow. 

*You may add more code cells if needed.*

```r
#Set seed for reproducibility
set.seed(30526602)

# Load tidymodels library
library(tidymodels) 

# Split the data (75% training, 25% testing) and remove rows with NA before predicting
wmm_split <- initial_split(wmm_tidy, prop = 0.75)
wmm_training <- training(wmm_split) |> 
    filter(!is.na(conf_wins_pct))
wmm_testing <- testing(wmm_split) |> 
    filter(!is.na(conf_wins_pct))

# Define a recipe to scale and center predictors
wmm_recipe <- recipe(elite_8 ~ seed + conf_wins_pct, data = wmm_training) |>
  step_center(all_predictors()) |>
  step_scale(all_predictors())

# Define the kNN model with tunable neighbors
knn_model_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
  set_engine("kknn") |>
  set_mode("classification")

# Define the workflow
knn_workflow <- workflow() |>
  add_recipe(wmm_recipe) |>
  add_model(knn_model_tune)

# Set up 5-fold cross-validation to evaluate the model's performance
wmm_folds <- vfold_cv(wmm_training, v = 5)

# Define tuning grid for k values
k_grid <- tibble(neighbors = 1:20)

# Tune k using CV
knn_tune <- knn_workflow |>
    tune_grid(resamples = wmm_folds, 
              grid = k_grid,
              metrics = metric_set(accuracy))

# Collect metrics from tuning results
knn_tune_results <- knn_tune |>
    collect_metrics()

# Select the row with the highest mean accuracy
best_k <- knn_tune_results |>
    filter(.metric == "accuracy") |>    # only the accuracy metrics
    arrange(desc(mean)) |>    # arrange in decending order by the mean accuracy
    slice(1)    # take the 1st or top row

# Get the best k-value
best_k_value <- best_k$neighbors
best_k_value    # print the best k-value

# Redefine the kNN model with best k
knn_model <- nearest_neighbor(weight_func = "rectangular", neighbors = best_k_value) |>
  set_engine("kknn") |>
  set_mode("classification")

# Rebuild the workflow with the model w/ the best k-value
knn_workflow <- workflow() |>
  add_recipe(wmm_recipe) |>
  add_model(knn_model)

# Fit the model to the whole training set
knn_fit <- knn_workflow |> 
    fit(data = wmm_training)

# Predict on test set
elite_8_predictions <- predict(knn_fit, wmm_testing, type = "class") |>     # pass fitted model & test dataset to predict()
  bind_cols(wmm_testing) |>    # add the columns of predictions to the OG test data
  mutate(elite_8 = as.factor(elite_8))   # make sure truth is a factor

# Evaluate model performance using our testing data

# Find out our model's accuracy, specify truth and estimate arguments
wmm_metrics <- elite_8_predictions |>
  accuracy(truth = elite_8, estimate = .pred_class)    # .pred_class is the predicted class from predict() and elite_8 is the true outcome
wmm_metrics    # print the accuracy

# Create a confusion matrix, specify truth and estimate arguments
wmm_conf_mat <- elite_8_predictions |>
  conf_mat(truth = elite_8, estimate = .pred_class)    # .pred_class is the predicted class from predict() and elite_8 is the true outcome
wmm_conf_mat    # print the confusion matrix
```
